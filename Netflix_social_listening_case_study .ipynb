```python
import pandas as pd
```

```python
# Loading a twitter dataset about Nextflix
df1 = pd.read_json('data/Netflix.jsonl', lines = True)
```

```python
# Showing the first few rows of the dataset
df1.head ()
```

```python
# Flatten the jsonl dataset and see how is it going
twitter_data = pd.json_normalize(df1['data'])
twitter_data.head()
```

```python
# Use the code provided in the tutorial to extract the twitter information out of the jsonl file.
def get_public_metrics(row):
    if 'public_metrics' in row.keys():
        if type(row['public_metrics']) == dict:
            for key, value in row['public_metrics'].items():
                row['metric_' + str(key)] = value
    return row

def get_tweets(df):
    if 'data' not in df.columns:
        return None
    results = pd.DataFrame()
    for item in df['data'].values.tolist():
        results = pd.concat([results, pd.DataFrame(item)])
        
    results = results.apply(get_public_metrics, axis=1)
        
    results = results.reset_index()
    del results['index']
        
    return results
```

```python
tweets = get_tweets(df1)
```

```python
# Now we can see more clearly about the twitter data.
tweets.head()
```

```python
# Understand what columns and what types of information are there in the dataset.
tweets.info()
```

```python
# Because the id datatype in the two dataset are differet we need to convert the 'id' column from object to int64
tweets['id'] = tweets['id'].astype('int64')

# Verify the conversion
print(tweets['id'].dtype)
```

```python
# Loading the correlated sentiment dataset.
df2 = pd.read_csv('data/Netflix_EN_withtext_completed.csv')
```

```python
df2.head()
```

```python
df2.info()
```

```python
# Because we only need information related to the tweets' sentiment and the engagement, and the basic information, we can make a minized dataset as below.
relevant_columns = ['id', 'text', 'lang', 'metric_retweet_count', 'metric_reply_count', 
                    'metric_like_count', 'metric_quote_count', 'metric_impression_count']
```

```python
minimized_df = tweets[relevant_columns]
```

```python
minimized_df.head()
```

```python
# Making a merged dataset containing both the twitter information and its sentiment information.
df3 = pd.merge(df2,minimized_df,on='id',how= 'inner')
```

```python
df3.info()
```

```python
df3.to_csv('merged_dataset.csv', index=False)
```

```python
# Creating one variable that operationalizes the sentiment.
# Calculating the sentiment level by adding up the three columns about sentiment, because this is a dataset provided by SentiStrength and adding up is easy.
df3['sentiment_level'] = df3['positive'] + df3['negative'] + df3['neutral']
```

```python
def classify_sentiment_level(value):
    if value > 0:
        return 'positive'
    elif value < 0:
        return 'negative'
    else:
        return 'neutral'
```

```python
df3['sentiment_category'] = df3['sentiment_level'].apply(classify_sentiment_level)
```

```python
df3[['sentiment_level', 'sentiment_category']].head()
```

```python
# Categorize tweets based on retweets & originial tweets.
```

```python
df3['tweet_type'] = df3['text_x'].apply(lambda x: 'retweet' if x.startswith('RT') else 'original')
```

```python
print (df3[['tweet_type','text_x']])
```

```python
# Create the 'retweet' column: 1 for retweet, 0 for original
# Display the first few rows to check the results
df3['retweet'] = df3['tweet_type'].apply(lambda x: 1 if x == 'retweet' else 0)
df3[['text_x', 'tweet_type', 'retweet']].head()
```

```python
# Load the manually checked validation sample used for sentiment reliability checking
mc_sample = pd.read_excel('validation/sample_checked.xlsx')

# Extract the corresponding automated sentiment labels from the main dataset
sentiment_data = df3[['text_x', 'sentiment_level']]

# Preview manually coded sample
mc_sample.head()
```

```python
# Merge automated sentiment labels with the manually checked sample
merged_data = pd.merge(sentiment_data, mc_sample[['text_x', 'sentiment_checked']], on='text_x', how='inner')

final_data = merged_data[['text_x', 'sentiment_checked', 'sentiment_level']]

final_data.head()
```

```python
# Calculate agreement between automated sentiment and manual sentiment

final_data['match'] = final_data['sentiment_checked'] == final_data['sentiment_level']

agreement_rate = final_data['match'].mean()

print(f"Agreement rate: {agreement_rate:.2%}")
```

```python
# Review mismatched cases

mismatches = final_data[final_data['match'] == False]

mismatches[['text_x', 'sentiment_checked', 'sentiment_level']].head(10)
```

Comparing the description of the two column, we can see they are the same and the inter-coder reliability is strong.

## Data visualization

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# What is the overall emotional tone of Netflix-related conversations?

```python
# What is the overall emotional tone of Netflix-related conversations?
plt.figure(figsize=(7, 5))

sns.countplot(

    data=df3,

    x='sentiment_category',

    order=['negative', 'neutral', 'positive']

)

plt.title('Distribution of Sentiment in Netflix-related Tweets')

plt.xlabel('Sentiment')

plt.ylabel('Number of Tweets')

plt.tight_layout()

plt.savefig('outputs/sentiment_distribution.png', dpi=300, bbox_inches='tight')

plt.show()
```

**Insight:** The sentiment distribution provides an overview of the emotional tone in Netflix-related conversations. This helps marketers understand whether audience discussion is mainly positive, neutral, or negative before evaluating engagement performance.

## Which sentiment group generates stronger audience interaction?

```python
engagement_cols = ['metric_like_count', 'metric_retweet_count', 'metric_reply_count', 'metric_quote_count']

df3['total_engagement'] = df3[engagement_cols].sum(axis=1)
```

```python
engagement_by_sentiment = (
    df3.groupby('sentiment_category')['total_engagement']
    .mean()
    .reindex(['negative', 'neutral', 'positive'])
    .reset_index()
)

plt.figure(figsize=(7, 5))

sns.barplot(
    data=engagement_by_sentiment,
    x='sentiment_category',
    y='total_engagement'
)

plt.title('Average Engagement by Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Average Total Engagement')
plt.tight_layout()
plt.savefig('outputs/engagement_by_sentiment.png',dpi=300,bbox_inches='tight')
plt.show()
```

**Insight:** Comparing average engagement by sentiment helps identify emotional positive posts are more likely to attract interaction. This is useful for social listening and campaign performance review.

#  Are original tweets and retweets showing different sentiment patterns?

```python
plt.figure(figsize=(8, 5))

sns.countplot(

    data=df3,

    x='sentiment_category',

    hue='tweet_type',

    order=['negative', 'neutral', 'positive']

)

plt.title('Sentiment Distribution by Tweet Type')

plt.xlabel('Sentiment')

plt.ylabel('Number of Tweets')

plt.legend(title='Tweet Type')

plt.tight_layout()
plt.savefig('outputs/tweet_type_sentiment.png',dpi=300,bbox_inches='tight')

plt.show()
```

**Insight:** Comparing original tweets and retweets adds context to audience behaviour. Retweets can reflect amplification of existing opinions, while original tweets may better indicate direct audience reactions.
